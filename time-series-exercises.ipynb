```python
response = requests.get('https://swapi.dev/api/people/')
data = response.json()

df = pd.DataFrame()

for i in range(10):
    response = requests.get(f'https://swapi.dev/api/people/?page={i}')
    data = response.json()
    if len(data.keys()) > 1:
        df = pd.concat([df, pd.DataFrame(data['results'])])
    
df = df.reset_index()
df = df.drop(columns = ['index'])
df.to_csv('people.csv')
df.shape
```
